```python
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
import matplotlib.pyplot as plt
import gc
from scipy import signal
from sklearn.metrics import mean_squared_error
import seaborn as sns

# ===================================================================
# SETUP
# ===================================================================
torch.manual_seed(42)
np.random.seed(42)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# ===================================================================
# DATA LOADING & PREPARATION (Unchanged from before)
# ===================================================================
print("\nLoading and CLEANING dataset for cGAN training...")
try:
    with np.load("/kaggle/input/largest-alzheimer-eeg-dataset/integrated_eeg_dataset.npz") as data:
        X_raw_full = data['X_raw']
        y_labels_str_full = data['y_labels'][:, 0]

    class_0_labels_str = ['0', '0.0']
    class_1_labels_str = ['1', '1.0']
    mask_c0 = np.isin(y_labels_str_full, class_0_labels_str)
    mask_c1 = np.isin(y_labels_str_full, class_1_labels_str)
    valid_mask = mask_c0 | mask_c1

    X_raw_clean = X_raw_full[valid_mask]
    y_labels_filtered = y_labels_str_full[valid_mask]
    
    y_clean = np.zeros(len(y_labels_filtered), dtype=int)
    y_clean[np.isin(y_labels_filtered, class_1_labels_str)] = 1
    
    print(f"Using {len(y_clean)} clean samples for cGAN training.")
    print(f"Class distribution - Class 0: {np.sum(y_clean == 0)}, Class 1: {np.sum(y_clean == 1)}")

except Exception as e:
    print(f"Error loading dataset: {e}")
    raise

print("Normalizing data...")
X_normalized = np.empty_like(X_raw_clean, dtype=np.float32)
for c in range(X_raw_clean.shape[2]):
    min_val = X_raw_clean[:, :, c].min()
    max_val = X_raw_clean[:, :, c].max()
    X_normalized[:, :, c] = 2 * ((X_raw_clean[:, :, c] - min_val) / (max_val - min_val)) - 1

del X_raw_full, X_raw_clean, y_labels_str_full, y_labels_filtered
gc.collect()

X_tensor = torch.from_numpy(X_normalized)
y_tensor = torch.from_numpy(y_clean).type(torch.LongTensor)

batch_size = 64
dataset = TensorDataset(X_tensor, y_tensor)
dataloader = DataLoader(
    dataset, batch_size=batch_size, shuffle=True,
    num_workers=4, pin_memory=True, persistent_workers=True
)

del X_normalized, X_tensor
gc.collect()
print("Prepared DataLoader for improved cGAN.")

# ===================================================================
# V5 MODELS: ResGenerator and SpectralNorm Discriminator
# ===================================================================
#temporal gan
class ResBlock(nn.Module):
    def __init__(self, in_channels, out_channels):
        super(ResBlock, self).__init__()
        self.conv_block = nn.Sequential(
            nn.BatchNorm1d(in_channels),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Upsample(scale_factor=2),
            nn.Conv1d(in_channels, out_channels, kernel_size=3, padding=1),
            nn.BatchNorm1d(out_channels),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Conv1d(out_channels, out_channels, kernel_size=3, padding=1),
        )
        self.shortcut = nn.Sequential(
            nn.Upsample(scale_factor=2),
            nn.Conv1d(in_channels, out_channels, kernel_size=1, padding=0)
        )
    def forward(self, x):
        return self.conv_block(x) + self.shortcut(x)

class ResGenerator(nn.Module):
    def __init__(self, latent_dim=100, num_classes=2, channels=19, timesteps=128):
        super(ResGenerator, self).__init__()
        self.timesteps = timesteps
        self.channels = channels
        self.label_emb = nn.Embedding(num_classes, latent_dim)
        self.fc = nn.Linear(latent_dim * 2, 256 * (timesteps // 8))
        self.res_blocks = nn.Sequential(
            ResBlock(256, 128),
            ResBlock(128, 64),
            ResBlock(64, 32),
        )
        self.final_conv = nn.Sequential(
            nn.BatchNorm1d(32),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Conv1d(32, channels, kernel_size=3, padding=1),
            nn.Tanh()
        )
    def forward(self, noise, labels):
        c = self.label_emb(labels)
        x = torch.cat([noise, c], 1)
        x = self.fc(x)
        x = x.reshape(x.size(0), 256, self.timesteps // 8)
        x = self.res_blocks(x)
        x = self.final_conv(x)
        return x.transpose(1, 2)

# --- Discriminator with Spectral Norm and .reshape() ---
from torch.nn.utils import spectral_norm
class ImprovedDiscriminator(nn.Module):
    def __init__(self, num_classes=2, timesteps=128, channels=19):
        super(ImprovedDiscriminator, self).__init__()
        self.timesteps = timesteps
        self.channels = channels
        self.label_emb = nn.Embedding(num_classes, timesteps * channels)
        self.conv_layers = nn.Sequential(
            spectral_norm(nn.Conv1d(channels, 64, kernel_size=5, stride=2, padding=2)),
            nn.LeakyReLU(0.2, inplace=True),
            spectral_norm(nn.Conv1d(64, 128, kernel_size=5, stride=2, padding=2)),
            nn.LeakyReLU(0.2, inplace=True),
            spectral_norm(nn.Conv1d(128, 256, kernel_size=5, stride=2, padding=2)),
            nn.LeakyReLU(0.2, inplace=True),
        )
        self._calculate_conv_output_size()
        self.classifier = nn.Sequential(
            spectral_norm(nn.Linear(self.conv_output_size + (timesteps * channels), 512)),
            nn.LeakyReLU(0.2, inplace=True),
            spectral_norm(nn.Linear(512, 256)),
            nn.LeakyReLU(0.2, inplace=True),
            spectral_norm(nn.Linear(256, 1))
        )

    def _calculate_conv_output_size(self):
        with torch.no_grad():
            dummy_input = torch.randn(1, self.channels, self.timesteps)
            dummy_output = self.conv_layers(dummy_input)
            self.conv_output_size = dummy_output.reshape(1, -1).size(1) # Use .reshape()

    def forward(self, eeg, labels):
        x = eeg.transpose(1, 2)
        x = self.conv_layers(x)
        x = x.reshape(x.size(0), -1)
        c = self.label_emb(labels).reshape(labels.size(0), -1) # Use .reshape()
        x = torch.cat([x, c], 1)
        return self.classifier(x)

# --- Helper functions (unchanged) ---
def compute_gradient_penalty(discriminator, real_samples, fake_samples, labels, device):
    alpha = torch.rand(real_samples.size(0), 1, 1).to(device)
    interpolates = (alpha * real_samples + (1 - alpha) * fake_samples).requires_grad_(True)
    d_interpolates = discriminator(interpolates, labels)
    fake = torch.ones(real_samples.size(0), 1).to(device)
    gradients = torch.autograd.grad(outputs=d_interpolates, inputs=interpolates, grad_outputs=fake, create_graph=True, retain_graph=True, only_inputs=True)[0]
    gradients = gradients.reshape(gradients.size(0), -1)
    gradient_penalty = ((gradients.norm(2, dim=1) - 1) ** 2).mean()
    return gradient_penalty

def evaluate_sample_quality(real_samples, generated_samples):
    metrics = {}
    sample_size = min(100, len(real_samples), len(generated_samples))
    real_subset = real_samples[np.random.choice(len(real_samples), sample_size, replace=False)]
    gen_subset = generated_samples[np.random.choice(len(generated_samples), sample_size, replace=False)]
    real_psds = [signal.welch(real_subset[i, :, 0], fs=128)[1] for i in range(sample_size)]
    gen_psds = [signal.welch(gen_subset[i, :, 0], fs=128)[1] for i in range(sample_size)]
    metrics['psd_mse'] = mean_squared_error(np.mean(real_psds, axis=0), np.mean(gen_psds, axis=0))
    return metrics

def plot_training_progress(g_losses, d_losses, quality_metrics, save_interval):
    fig, axes = plt.subplots(1, 2, figsize=(18, 6))
    axes[0].plot(g_losses, label='Generator Loss')
    axes[0].plot(d_losses, label='Discriminator Loss')
    axes[0].set_title('Training Losses')
    axes[0].set_xlabel('Epoch')
    axes[0].legend(); axes[0].grid(True)
    if quality_metrics:
        epochs = np.arange(1, len(quality_metrics) + 1) * save_interval
        psd_mses = [m['psd_mse'] for m in quality_metrics]
        axes[1].plot(epochs, psd_mses, 'r-o', label='PSD MSE')
        axes[1].set_title('Sample Quality (Lower is Better)')
        axes[1].set_xlabel('Epoch')
        axes[1].set_ylabel('PSD MSE')
        axes[1].legend(); axes[1].grid(True)
    plt.tight_layout()
    plt.savefig('improved_training_progress.png')
    plt.show()

def compare_real_vs_generated(real_samples, generated_samples, num_samples=3):
    fig, axes = plt.subplots(num_samples, 2, figsize=(15, num_samples * 4))
    fig.suptitle('Visual Comparison of Real vs. Generated Samples', fontsize=16)
    for i in range(num_samples):
        real_idx, gen_idx = np.random.randint(0, len(real_samples)), np.random.randint(0, len(generated_samples))
        axes[i, 0].plot(real_samples[real_idx, :, 0], label='Real')
        axes[i, 0].plot(generated_samples[gen_idx, :, 0], label='Generated', alpha=0.8)
        axes[i, 0].set_title(f'Time Series (Sample {i+1})'); axes[i, 0].legend()
        f_real, psd_real = signal.welch(real_samples[real_idx, :, 0], fs=128)
        f_gen, psd_gen = signal.welch(generated_samples[gen_idx, :, 0], fs=128)
        axes[i, 1].semilogy(f_real, psd_real, label='Real')
        axes[i, 1].semilogy(f_gen, psd_gen, label='Generated', alpha=0.8)
        axes[i, 1].set_title(f'Power Spectral Density (Sample {i+1})'); axes[i, 1].legend()
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.savefig('real_vs_generated_comparison.png')
    plt.show()

# --- CORRECTED Training Function ---
def train_v5_gan(dataloader, num_epochs=250, critic_iterations=5, save_interval=20):
    generator = ResGenerator().to(device)
    discriminator = ImprovedDiscriminator().to(device)
    
    lr_g, lr_d = 2e-4, 1e-4
    g_optimizer = optim.Adam(generator.parameters(), lr=lr_g, betas=(0.5, 0.999))
    d_optimizer = optim.Adam(discriminator.parameters(), lr=lr_d, betas=(0.5, 0.999))
    
    g_scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(g_optimizer, 'min', factor=0.5, patience=15, verbose=True)
    d_scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(d_optimizer, 'min', factor=0.5, patience=15, verbose=True)
    
    g_losses, d_losses, quality_metrics = [], [], []

    # GP Annealing parameters
    lambda_gp_initial, lambda_gp_final = 0.1, 10
    anneal_epochs = 50
    
    print("\nStarting V5  cWGAN-GP training with ResNet generator...")
    for epoch in range(num_epochs):
        # GP annealing logic
        if epoch < anneal_epochs:
            lambda_gp = lambda_gp_initial + (lambda_gp_final - lambda_gp_initial) * (epoch / anneal_epochs)
        else:
            lambda_gp = lambda_gp_final
        # *** FIX #1: Initialize epoch loss lists ***
        epoch_g_losses = []
        epoch_d_losses = []
        
        for i, (real_eegs, labels) in enumerate(dataloader):
            batch_size = real_eegs.size(0)
            real_eegs, labels = real_eegs.to(device), labels.to(device)
            
            d_optimizer.zero_grad()
            z = torch.randn(batch_size, 100).to(device)
            fake_eegs = generator(z, labels).detach()
            real_validity = discriminator(real_eegs, labels)
            fake_validity = discriminator(fake_eegs, labels)
            gradient_penalty = compute_gradient_penalty(discriminator, real_eegs.data, fake_eegs.data, labels.data, device)
            d_loss = -torch.mean(real_validity) + torch.mean(fake_validity) + lambda_gp * gradient_penalty
            d_loss.backward()
            d_optimizer.step()
            epoch_d_losses.append(d_loss.item()) # Append batch loss

            if i % critic_iterations == 0:
                g_optimizer.zero_grad()
                z = torch.randn(batch_size, 100).to(device)
                gen_labels = torch.randint(0, 2, (batch_size,)).to(device)
                gen_eegs = generator(z, gen_labels)
                g_loss = -torch.mean(discriminator(gen_eegs, gen_labels))
                g_loss.backward()
                g_optimizer.step()
                epoch_g_losses.append(g_loss.item()) # Append batch loss

        # *** FIX #1 (cont.): Calculate average loss for the epoch ***
        avg_g_loss = np.mean(epoch_g_losses)
        avg_d_loss = np.mean(epoch_d_losses)
        g_losses.append(avg_g_loss)
        d_losses.append(avg_d_loss)
        
        g_scheduler.step(avg_g_loss)
        d_scheduler.step(avg_d_loss)

        if (epoch + 1) % save_interval == 0:
            print(f"Epoch [{epoch+1}/{num_epochs}], D Loss: {avg_d_loss:.4f}, G Loss: {avg_g_loss:.4f} , Lambda GP : {lambda_gp:.2f}")
            # *** FIX #2: Move LR print statement here ***
            print(f"  Learning Rates - G: {g_optimizer.param_groups[0]['lr']:.6f}, D: {d_optimizer.param_groups[0]['lr']:.6f}")
            with torch.no_grad():
                generator.eval()
                z_eval = torch.randn(500, 100).to(device)
                labels_eval = torch.randint(0, 2, (500,)).to(device)
                generated_samples = generator(z_eval, labels_eval).cpu().numpy()
                real_samples = dataloader.dataset.tensors[0][:500].numpy()
                metrics = evaluate_sample_quality(real_samples, generated_samples)
                quality_metrics.append(metrics)
                print(f"  Quality - PSD MSE: {metrics['psd_mse']:.8f}")
                generator.train()

    return generator, g_losses, d_losses, quality_metrics

# ===================================================================
# EXECUTION AND SAMPLE GENERATION
# ===================================================================
# Train the new improved model
trained_generator, g_losses, d_losses, quality_metrics = train_v5_gan(dataloader, num_epochs=250,  save_interval=20)

# Plot the training progress and sample quality
plot_training_progress(g_losses, d_losses, quality_metrics, save_interval=20)

# Visualize a comparison
print("\nVisual comparison of Real vs. Generated Samples:")
trained_generator.eval()
with torch.no_grad():
    z_vis = torch.randn(10, 100).to(device)
    labels_vis = torch.randint(0, 2, (10,)).to(device)
    vis_generated = trained_generator(z_vis, labels_vis).cpu().numpy()
vis_real = dataloader.dataset.tensors[0][:10].numpy()
compare_real_vs_generated(vis_real, vis_generated)

# Generate the final synthetic dataset for the next script
print("\nGenerating final synthetic dataset for augmentation...")
y_clean_final = dataloader.dataset.tensors[1].numpy()
class_counts = np.bincount(y_clean_final)
max_count = np.max(class_counts)
samples_to_generate = max_count - class_counts

synthetic_samples_list, synthetic_labels_list = [], []
for class_label, num_to_gen in enumerate(samples_to_generate):
    if num_to_gen > 0:
        print(f"Generating {num_to_gen} synthetic samples for class {class_label}...")
        with torch.no_grad():
            trained_generator.eval()
            # Generate in batches to avoid OOM on GPU
            generated_data_parts = []
            gen_batch_size = 512
            for i in range(0, num_to_gen, gen_batch_size):
                current_batch_size = min(gen_batch_size, num_to_gen - i)
                z_gen = torch.randn(current_batch_size, 100).to(device)
                labels_gen = torch.LongTensor([class_label] * current_batch_size).to(device)
                generated_data_parts.append(trained_generator(z_gen, labels_gen).cpu().numpy())
            generated_data = np.concatenate(generated_data_parts, axis=0)
        synthetic_samples_list.append(generated_data)
        generated_labels = np.zeros((num_to_gen, 3), dtype=object)
        generated_labels[:, 0] = str(class_label)
        generated_labels[:, 1] = -1
        generated_labels[:, 2] = "synthetic"
        synthetic_labels_list.append(generated_labels)

# Save the final augmented dataset file
if synthetic_samples_list:
    X_synthetic = np.concatenate(synthetic_samples_list, axis=0)
    y_synthetic = np.concatenate(synthetic_labels_list, axis=0)
    with np.load("/kaggle/input/largest-alzheimer-eeg-dataset/integrated_eeg_dataset.npz") as data:
        X_original, y_original = data['X_raw'], data['y_labels']
    save_path = "/kaggle/working/augmented_eeg_dataset.npz"
    np.savez_compressed(save_path, X_original=X_original, y_original=y_original, X_synthetic=X_synthetic, y_synthetic=y_synthetic)
    print(f"\nImproved augmented dataset saved successfully to {save_path}")
```

```python
# REVISED SCRIPT 2 (Classification) - WITH EEGNetFusion

import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.signal import welch
from scipy.integrate import simpson as simps
import gc

# ===================================================================
# SETUP
# ===================================================================
torch.manual_seed(42)
np.random.seed(42)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# ===================================================================
# FEATURE EXTRACTION (This part remains the same)
# ===================================================================
def extract_enhanced_features(X_batch, sfreq=128):
    # ... (Your existing, correct function)
    bands = {'delta': (0.5, 4), 'theta': (4, 8), 'alpha': (8, 13), 'beta': (13, 30), 'gamma': (30, 50)}
    nperseg = X_batch.shape[1]
    freqs, psd = welch(X_batch, fs=sfreq, nperseg=nperseg, axis=1)
    psd = np.transpose(psd, (0, 2, 1))
    all_band_powers = []
    freq_res = freqs[1] - freqs[0]
    for band_limits in bands.values():
        band_idx = np.logical_and(freqs >= band_limits[0], freqs <= band_limits[1])
        bp = simps(psd[:, :, band_idx], dx=freq_res, axis=-1)
        all_band_powers.append(bp)
    abs_bp = np.stack(all_band_powers, axis=-1)
    total_power = abs_bp.sum(axis=-1, keepdims=True)
    total_power[total_power == 0] = 1e-10
    rel_bp = abs_bp / total_power
    features = np.concatenate([
        rel_bp.reshape(X_batch.shape[0], -1),
        abs_bp.reshape(X_batch.shape[0], -1),
        X_batch.mean(axis=1),
        X_batch.std(axis=1),
        X_batch.max(axis=1),
        X_batch.min(axis=1)
    ], axis=1)
    return features

def extract_features_in_batches(X_raw, batch_size=1024):
    # ... (Your existing function)
    n_samples = X_raw.shape[0]
    all_features = []
    print(f"Extracting features for {n_samples} samples...")
    for i in range(0, n_samples, batch_size):
        X_batch = X_raw[i:min(i + batch_size, n_samples)]
        features_batch = extract_enhanced_features(X_batch)
        all_features.append(features_batch)
    return np.concatenate(all_features, axis=0)

# ===================================================================
# NEW EEGNetFusion MODEL DEFINITION (Copy the class from above here)
# ===================================================================
class EEGNetFusion(nn.Module):
    # ... (Paste the full class definition from section 1 here)
    def __init__(self, num_channels=19, num_timesteps=128, num_features=266, num_classes=2, F1=8, D=2, F2=16):
        super(EEGNetFusion, self).__init__()
        self.eegnet_branch = nn.Sequential(
            nn.Conv2d(1, F1, kernel_size=(1, 64), padding='same', bias=False),
            nn.BatchNorm2d(F1),
            nn.Conv2d(F1, F1 * D, kernel_size=(num_channels, 1), groups=F1, bias=False),
            nn.BatchNorm2d(F1 * D), nn.ELU(), nn.AvgPool2d(kernel_size=(1, 4)), nn.Dropout(p=0.5),
            nn.Conv2d(F1 * D, F2, kernel_size=(1, 16), padding='same', groups=F1*D, bias=False),
            nn.Conv2d(F2, F2, kernel_size=(1, 1), bias=False),
            nn.BatchNorm2d(F2), nn.ELU(), nn.AvgPool2d(kernel_size=(1, 8)), nn.Dropout(p=0.5),
            nn.Flatten()
        )
        self.feature_branch = nn.Sequential(
            nn.Linear(num_features, 128), nn.BatchNorm1d(128), nn.ELU(), nn.Dropout(p=0.5),
            nn.Linear(128, 64), nn.BatchNorm1d(64), nn.ELU()
        )
        with torch.no_grad():
            dummy_raw_input = torch.randn(1, 1, num_channels, num_timesteps)
            eegnet_out_size = self.eegnet_branch(dummy_raw_input).shape[1]
        self.classifier = nn.Sequential(
            nn.Linear(eegnet_out_size + 64, 128), nn.ELU(), nn.Dropout(p=0.5),
            nn.Linear(128, num_classes)
        )
    def forward(self, raw_eeg, features):
        raw_eeg_input = raw_eeg.permute(0, 2, 1).unsqueeze(1)
        eegnet_output = self.eegnet_branch(raw_eeg_input)
        feature_output = self.feature_branch(features)
        fused_output = torch.cat((eegnet_output, feature_output), dim=1)
        return self.classifier(fused_output)


# ===================================================================
# UPDATED TRAINING AND EVALUATION FUNCTIONS
# ===================================================================
def train_model(model, train_loader, test_loader, num_epochs=50, model_save_path='best_model.pth'):
    model = model.to(device)
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=0.001, weight_decay=1e-4)
    scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, 'max', factor=0.2, patience=5, verbose=True)
    best_acc = 0.0
    
    for epoch in range(num_epochs):
        model.train()
        for raw_eeg_batch, features_batch, labels_batch in train_loader:
            raw_eeg_batch, features_batch, labels_batch = raw_eeg_batch.to(device), features_batch.to(device), labels_batch.to(device)
            optimizer.zero_grad()
            # *** Pass both inputs to the model ***
            outputs = model(raw_eeg_batch, features_batch)
            loss = criterion(outputs, labels_batch)
            loss.backward()
            optimizer.step()
        
        val_acc = evaluate_accuracy(model, test_loader)
        scheduler.step(val_acc)
        if val_acc > best_acc:
            best_acc = val_acc
            torch.save(model.state_dict(), model_save_path)
        if (epoch + 1) % 5 == 0:
            print(f"Epoch {epoch+1}/{num_epochs}, Val Acc: {val_acc:.4f}")
    
    print(f"Training finished. Best validation accuracy: {best_acc:.4f}")
    model.load_state_dict(torch.load(model_save_path))
    return model

def evaluate_accuracy(model, loader):
    model.eval()
    correct, total = 0, 0
    with torch.no_grad():
        for raw_eeg_batch, features_batch, labels_batch in loader:
            raw_eeg_batch, features_batch, labels_batch = raw_eeg_batch.to(device), features_batch.to(device), labels_batch.to(device)
            outputs = model(raw_eeg_batch, features_batch)
            _, predicted = torch.max(outputs.data, 1)
            total += labels_batch.size(0)
            correct += (predicted == labels_batch).sum().item()
    return correct / total

def full_evaluation(model, loader, dataset_name):
    model.eval()
    all_preds, all_labels = [], []
    with torch.no_grad():
        for raw_eeg_batch, features_batch, labels_batch in loader:
            raw_eeg_batch, features_batch, labels_batch = raw_eeg_batch.to(device), features_batch.to(device), labels_batch.to(device)
            outputs = model(raw_eeg_batch, features_batch)
            _, preds = torch.max(outputs, 1)
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels_batch.cpu().numpy())
    print(f"\n--- Full Evaluation on {dataset_name} ---")
    print(classification_report(all_labels, all_preds, target_names=['Class 0', 'Class 1']))
    cm = confusion_matrix(all_labels, all_preds)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues').set_title(f'Confusion Matrix ({dataset_name})')
    plt.show()

# ===================================================================
# UPDATED MAIN EXECUTION PIPELINE
# ===================================================================
def process_and_prepare_dataset(X_raw, y_labels):
    """Helper function to process raw data into tensors for the new model."""
    print("Extracting handcrafted features...")
    X_features = extract_features_in_batches(X_raw)
    
    # Normalize the raw EEG signals for the EEGNet branch
    print("Normalizing raw EEG signals...")
    X_raw_normalized = np.empty_like(X_raw, dtype=np.float32)
    for i in range(X_raw.shape[0]):
        for c in range(X_raw.shape[2]):
            X_raw_normalized[i, :, c] = (X_raw[i, :, c] - X_raw[i, :, c].mean()) / (X_raw[i, :, c].std() + 1e-8)
            
    # Convert all to tensors
    X_raw_tensor = torch.FloatTensor(X_raw_normalized)
    X_features_tensor = torch.FloatTensor(X_features)
    y_tensor = torch.LongTensor(y_labels)
    
    return X_raw_tensor, X_features_tensor, y_tensor


if __name__ == "__main__":
    original_dataset_path = "/kaggle/input/largest-alzheimer-eeg-dataset/integrated_eeg_dataset.npz"
    augmented_dataset_path = "/kaggle/working/augmented_eeg_dataset.npz"

    # --- Step 1: Prepare and Train on Original Data ---
    print("\n" + "="*50 + "\nSTEP 1: PROCESSING ORIGINAL DATASET\n" + "="*50)
    
    with np.load(original_dataset_path) as data:
        X_orig_raw, y_orig_str = data['X_raw'], data['y_labels'][:, 0]
    
    # Clean the data (same logic as before)
    valid_mask = np.isin(y_orig_str, ['0', '0.0', '1', '1.0'])
    X_orig_clean, y_orig_str_clean = X_orig_raw[valid_mask], y_orig_str[valid_mask]
    y_orig_clean = np.isin(y_orig_str_clean, ['1', '1.0']).astype(int)
    
    X_orig_raw_t, X_orig_feat_t, y_orig_t = process_and_prepare_dataset(X_orig_clean, y_orig_clean)
    
    # Create dataset with THREE tensors now
    orig_dataset = TensorDataset(X_orig_raw_t, X_orig_feat_t, y_orig_t)
    
    # Split and create DataLoaders
    train_indices, test_indices = train_test_split(range(len(orig_dataset)), test_size=0.2, random_state=42, stratify=y_orig_t)
    orig_train_dataset = torch.utils.data.Subset(orig_dataset, train_indices)
    orig_test_dataset = torch.utils.data.Subset(orig_dataset, test_indices)
    
    orig_train_loader = DataLoader(orig_train_dataset, batch_size=64, shuffle=True)
    orig_test_loader = DataLoader(orig_test_dataset, batch_size=64, shuffle=False)
    
    # Train the model
    model_orig = EEGNetFusion()
    print("\nTraining EEGNetFusion model on original data...")
    model_orig = train_model(model_orig, orig_train_loader, orig_test_loader, num_epochs=50, model_save_path='model_orig.pth')

    # --- Step 2: Prepare and Train on Augmented Data ---
    print("\n" + "="*50 + "\nSTEP 2: PROCESSING AUGMENTED DATASET\n" + "="*50)

    with np.load(augmented_dataset_path, allow_pickle=True) as data:
        X_aug_raw = np.concatenate((data['X_original'], data['X_synthetic']), axis=0)
        y_aug_str = np.concatenate((data['y_original'][:, 0], data['y_synthetic'][:, 0]), axis=0)
        
    valid_mask_aug = np.isin(y_aug_str, ['0', '0.0', '1', '1.0'])
    X_aug_clean, y_aug_str_clean = X_aug_raw[valid_mask_aug], y_aug_str[valid_mask_aug]
    y_aug_clean = np.isin(y_aug_str_clean, ['1', '1.0']).astype(int)
    
    X_aug_raw_t, X_aug_feat_t, y_aug_t = process_and_prepare_dataset(X_aug_clean, y_aug_clean)
    
    aug_dataset = TensorDataset(X_aug_raw_t, X_aug_feat_t, y_aug_t)
    
    train_indices_aug, test_indices_aug = train_test_split(range(len(aug_dataset)), test_size=0.2, random_state=42, stratify=y_aug_t)
    aug_train_dataset = torch.utils.data.Subset(aug_dataset, train_indices_aug)
    aug_test_dataset = torch.utils.data.Subset(aug_dataset, test_indices_aug)
    
    aug_train_loader = DataLoader(aug_train_dataset, batch_size=64, shuffle=True)
    aug_test_loader = DataLoader(aug_test_dataset, batch_size=64, shuffle=False)
    
    model_aug = EEGNetFusion()
    print("\nTraining EEGNetFusion model on augmented data...")
    model_aug = train_model(model_aug, aug_train_loader, aug_test_loader, num_epochs=50, model_save_path='model_aug.pth')

    # --- Step 3: Final Evaluation ---
    print("\n" + "="*50 + "\nSTEP 3: FINAL COMPARATIVE EVALUATION\n" + "="*50)
    full_evaluation(model_orig, orig_test_loader, "Original Model on Original Test Set")
    full_evaluation(model_aug, aug_test_loader, "Augmented Model on Augmented Test Set")
```
