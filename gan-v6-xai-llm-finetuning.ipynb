```python
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
import matplotlib.pyplot as plt
import gc
from scipy import signal
from sklearn.metrics import mean_squared_error
import seaborn as sns
from torch.nn.utils import spectral_norm

# ===================================================================
# SETUP
# ===================================================================
torch.manual_seed(42)
np.random.seed(42)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# ===================================================================
# DATA LOADING & PREPARATION (Unchanged from before)
# ===================================================================
print("\nLoading and CLEANING dataset for cGAN training...")
try:
    with np.load("/kaggle/input/adftd-alzheimer-eeg-only/adftd_complete.npz") as data:
        X_raw_full = data['X_raw']
        y_labels_str_full = data['y_labels'][:, 0]

    class_0_labels_str = ['0', '0.0']
    class_1_labels_str = ['1', '1.0']
    mask_c0 = np.isin(y_labels_str_full, class_0_labels_str)
    mask_c1 = np.isin(y_labels_str_full, class_1_labels_str)
    valid_mask = mask_c0 | mask_c1

    X_raw_clean = X_raw_full[valid_mask]
    y_labels_filtered = y_labels_str_full[valid_mask]
    
    y_clean = np.zeros(len(y_labels_filtered), dtype=int)
    y_clean[np.isin(y_labels_filtered, class_1_labels_str)] = 1
    
    print(f"Using {len(y_clean)} clean samples for cGAN training.")
    print(f"Class distribution - Class 0: {np.sum(y_clean == 0)}, Class 1: {np.sum(y_clean == 1)}")

except Exception as e:
    print(f"Error loading dataset: {e}")
    raise

print("Normalizing data...")
X_normalized = np.empty_like(X_raw_clean, dtype=np.float32)
for c in range(X_raw_clean.shape[2]):
    min_val = X_raw_clean[:, :, c].min()
    max_val = X_raw_clean[:, :, c].max()
    X_normalized[:, :, c] = 2 * ((X_raw_clean[:, :, c] - min_val) / (max_val - min_val)) - 1

del X_raw_full, X_raw_clean, y_labels_str_full, y_labels_filtered
gc.collect()

X_tensor = torch.from_numpy(X_normalized)
y_tensor = torch.from_numpy(y_clean).type(torch.LongTensor)

batch_size = 64
dataset = TensorDataset(X_tensor, y_tensor)
dataloader = DataLoader(
    dataset, batch_size=batch_size, shuffle=True,
    num_workers=4, pin_memory=True, persistent_workers=True
)

del X_normalized, X_tensor
gc.collect()
print("Prepared DataLoader for improved cGAN.")

# =============================================================================
# V6 MODELS: ResGenerator and SpectralNorm Discriminator with Noise Injection
# =============================================================================
#temporal gan
class ResBlock(nn.Module):
    def __init__(self, in_channels, out_channels):
        super(ResBlock, self).__init__()
        self.conv_block = nn.Sequential(
            nn.BatchNorm1d(in_channels),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Upsample(scale_factor=2),
            nn.Conv1d(in_channels, out_channels, kernel_size=3, padding=1),
            nn.BatchNorm1d(out_channels),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Conv1d(out_channels, out_channels, kernel_size=3, padding=1),
        )
        self.shortcut = nn.Sequential(
            nn.Upsample(scale_factor=2),
            nn.Conv1d(in_channels, out_channels, kernel_size=1, padding=0)
        )
    def forward(self, x):
        return self.conv_block(x) + self.shortcut(x)

class ResGenerator(nn.Module):
    def __init__(self, latent_dim=100, num_classes=2, channels=19, timesteps=128):
        super(ResGenerator, self).__init__()
        self.timesteps = timesteps
        self.channels = channels
        self.label_emb = nn.Embedding(num_classes, latent_dim)
        self.fc = nn.Linear(latent_dim * 2, 256 * (timesteps // 8))
        self.res_blocks = nn.Sequential(
            ResBlock(256, 128),
            ResBlock(128, 64),
            ResBlock(64, 32),
        )
        self.final_conv = nn.Sequential(
            nn.BatchNorm1d(32),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Conv1d(32, channels, kernel_size=3, padding=1),
            nn.Tanh()
        )
    def forward(self, noise, labels):
        c = self.label_emb(labels)
        x = torch.cat([noise, c], 1)
        x = self.fc(x)
        x = x.reshape(x.size(0), 256, self.timesteps // 8)
        x = self.res_blocks(x)
        x = self.final_conv(x)
        return x.transpose(1, 2)

# --- Discriminator with Spectral Norm and .reshape() ---
from torch.nn.utils import spectral_norm
class ImprovedDiscriminator(nn.Module):
    def __init__(self, num_classes=2, timesteps=128, channels=19):
        super(ImprovedDiscriminator, self).__init__()
        self.timesteps = timesteps
        self.channels = channels
        self.label_emb = nn.Embedding(num_classes, timesteps * channels)
        self.conv_layers = nn.Sequential(
            spectral_norm(nn.Conv1d(channels, 64, kernel_size=5, stride=2, padding=2)),
            nn.LeakyReLU(0.2, inplace=True),
            spectral_norm(nn.Conv1d(64, 128, kernel_size=5, stride=2, padding=2)),
            nn.LeakyReLU(0.2, inplace=True),
            spectral_norm(nn.Conv1d(128, 256, kernel_size=5, stride=2, padding=2)),
            nn.LeakyReLU(0.2, inplace=True),
        )
        self._calculate_conv_output_size()
        self.classifier = nn.Sequential(
            spectral_norm(nn.Linear(self.conv_output_size + (timesteps * channels), 512)),
            nn.LeakyReLU(0.2, inplace=True),
            spectral_norm(nn.Linear(512, 256)),
            nn.LeakyReLU(0.2, inplace=True),
            spectral_norm(nn.Linear(256, 1))
        )

    def _calculate_conv_output_size(self):
        with torch.no_grad():
            dummy_input = torch.randn(1, self.channels, self.timesteps)
            dummy_output = self.conv_layers(dummy_input)
            self.conv_output_size = dummy_output.reshape(1, -1).size(1) # Use .reshape()

    def forward(self, eeg, labels):
        # Diversity enhancement : Noise Injection ***
        if self.training:
            noise = torch.randn_like(eeg) * 0.05 #adding 5% noise
            eeg = eeg + noise
            
        x = eeg.transpose(1, 2)
        x = self.conv_layers(x)
        x = x.reshape(x.size(0), -1)
        c = self.label_emb(labels).reshape(labels.size(0), -1) # Use .reshape()
        x = torch.cat([x, c], 1)
        return self.classifier(x)

# --- Helper functions (unchanged) ---
def compute_gradient_penalty(discriminator, real_samples, fake_samples, labels, device):
    alpha = torch.rand(real_samples.size(0), 1, 1).to(device)
    interpolates = (alpha * real_samples + (1 - alpha) * fake_samples).requires_grad_(True)
    d_interpolates = discriminator(interpolates, labels)
    fake = torch.ones(real_samples.size(0), 1).to(device)
    gradients = torch.autograd.grad(outputs=d_interpolates, inputs=interpolates, grad_outputs=fake, create_graph=True, retain_graph=True, only_inputs=True)[0]
    gradients = gradients.reshape(gradients.size(0), -1)
    gradient_penalty = ((gradients.norm(2, dim=1) - 1) ** 2).mean()
    return gradient_penalty

def evaluate_sample_quality(real_samples, generated_samples):
    metrics = {}
    sample_size = min(100, len(real_samples), len(generated_samples))
    real_subset = real_samples[np.random.choice(len(real_samples), sample_size, replace=False)]
    gen_subset = generated_samples[np.random.choice(len(generated_samples), sample_size, replace=False)]
    real_psds = [signal.welch(real_subset[i, :, 0], fs=128)[1] for i in range(sample_size)]
    gen_psds = [signal.welch(gen_subset[i, :, 0], fs=128)[1] for i in range(sample_size)]
    metrics['psd_mse'] = mean_squared_error(np.mean(real_psds, axis=0), np.mean(gen_psds, axis=0))
    return metrics

def plot_training_progress(g_losses, d_losses, quality_metrics, save_interval):
    fig, axes = plt.subplots(1, 2, figsize=(18, 6))
    axes[0].plot(g_losses, label='Generator Loss')
    axes[0].plot(d_losses, label='Discriminator Loss')
    axes[0].set_title('Training Losses')
    axes[0].set_xlabel('Epoch')
    axes[0].legend(); axes[0].grid(True)
    if quality_metrics:
        epochs = np.arange(1, len(quality_metrics) + 1) * save_interval
        psd_mses = [m['psd_mse'] for m in quality_metrics]
        axes[1].plot(epochs, psd_mses, 'r-o', label='PSD MSE')
        axes[1].set_title('Sample Quality (Lower is Better)')
        axes[1].set_xlabel('Epoch')
        axes[1].set_ylabel('PSD MSE')
        axes[1].legend(); axes[1].grid(True)
    plt.tight_layout()
    plt.savefig('improved_training_progress.png')
    plt.show()

def compare_real_vs_generated(real_samples, generated_samples, num_samples=3):
    fig, axes = plt.subplots(num_samples, 2, figsize=(15, num_samples * 4))
    fig.suptitle('Visual Comparison of Real vs. Generated Samples', fontsize=16)
    for i in range(num_samples):
        real_idx, gen_idx = np.random.randint(0, len(real_samples)), np.random.randint(0, len(generated_samples))
        axes[i, 0].plot(real_samples[real_idx, :, 0], label='Real')
        axes[i, 0].plot(generated_samples[gen_idx, :, 0], label='Generated', alpha=0.8)
        axes[i, 0].set_title(f'Time Series (Sample {i+1})'); axes[i, 0].legend()
        f_real, psd_real = signal.welch(real_samples[real_idx, :, 0], fs=128)
        f_gen, psd_gen = signal.welch(generated_samples[gen_idx, :, 0], fs=128)
        axes[i, 1].semilogy(f_real, psd_real, label='Real')
        axes[i, 1].semilogy(f_gen, psd_gen, label='Generated', alpha=0.8)
        axes[i, 1].set_title(f'Power Spectral Density (Sample {i+1})'); axes[i, 1].legend()
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.savefig('real_vs_generated_comparison.png')
    plt.show()

# ======================================================================================
# V6 Training function with Mode seeking Loss
#+===================================================================================================

def train_v6_gan(dataloader, num_epochs=300, critic_iterations=5, save_interval=20):
    generator = ResGenerator().to(device)
    discriminator = ImprovedDiscriminator().to(device)
    
    lr_g, lr_d = 1e-4, 1e-4
    diversity_weight = 0.1
    g_optimizer = optim.Adam(generator.parameters(), lr=lr_g, betas=(0.5, 0.999))
    d_optimizer = optim.Adam(discriminator.parameters(), lr=lr_d, betas=(0.5, 0.999))
    
    g_scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(g_optimizer, 'min', factor=0.5, patience=15, verbose=True)
    d_scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(d_optimizer, 'min', factor=0.5, patience=15, verbose=True)
    
    g_losses, d_losses, quality_metrics = [], [], []

    # GP Annealing parameters
    lambda_gp_initial, lambda_gp_final = 0.1, 10
    anneal_epochs = 50
    
    print("\nStarting V5  cWGAN-GP training with ResNet generator...")
    for epoch in range(num_epochs):
        # GP annealing logic
        if epoch < anneal_epochs:
            lambda_gp = lambda_gp_initial + (lambda_gp_final - lambda_gp_initial) * (epoch / anneal_epochs)
        else:
            lambda_gp = lambda_gp_final
        # *** FIX #1: Initialize epoch loss lists ***
        epoch_g_losses = []
        epoch_d_losses = []
        
        for i, (real_eegs, labels) in enumerate(dataloader):
            batch_size = real_eegs.size(0)
            real_eegs, labels = real_eegs.to(device), labels.to(device)
            
            d_optimizer.zero_grad()
            z = torch.randn(batch_size, 100).to(device)
            fake_eegs = generator(z, labels).detach()
            real_validity = discriminator(real_eegs, labels)
            fake_validity = discriminator(fake_eegs, labels)
            gradient_penalty = compute_gradient_penalty(discriminator, real_eegs.data, fake_eegs.data, labels.data, device)
            d_loss = -torch.mean(real_validity) + torch.mean(fake_validity) + lambda_gp * gradient_penalty
            d_loss.backward()
            d_optimizer.step()
            epoch_d_losses.append(d_loss.item()) # Append batch loss

            if i % critic_iterations == 0:
                g_optimizer.zero_grad()
                z1 = torch.randn(batch_size, 100).to(device)
                gen_labels = torch.randint(0, 2, (batch_size,)).to(device)
                fake_eegs1 = generator(z, gen_labels)
                g_loss_adv = -torch.mean(discriminator(fake_eegs1, gen_labels))

                # Diversity Enhancement : Mode seeking Loss
                z2 = torch.randn(batch_size , 100).to(device)
                fake_eegs2 = generator(z2, gen_labels)
                lz = torch.mean(torch.abs(z2 - z1))
                lf = torch.mean(torch.abs(fake_eegs2 - fake_eegs1))
                g_loss_div = diversity_weight * (lz / (lf + 1e-8))

                g_loss = g_loss_adv + g_loss_div
                g_loss.backward()
                g_optimizer.step()
                epoch_g_losses.append(g_loss.item()) # Append batch loss

        # *** FIX #1 (cont.): Calculate average loss for the epoch ***
        avg_g_loss = np.mean(epoch_g_losses) if epoch_g_losses else 0
        avg_d_loss = np.mean(epoch_d_losses)
        g_losses.append(avg_g_loss)
        d_losses.append(avg_d_loss)
        
        g_scheduler.step(avg_g_loss)
        d_scheduler.step(avg_d_loss)

        if (epoch + 1) % save_interval == 0:
            print(f"Epoch [{epoch+1}/{num_epochs}], D Loss: {avg_d_loss:.4f}, G Loss: {avg_g_loss:.4f} , Lambda GP : {lambda_gp:.2f}")
            # *** FIX #2: Move LR print statement here ***
            print(f"  Learning Rates - G: {g_optimizer.param_groups[0]['lr']:.6f}, D: {d_optimizer.param_groups[0]['lr']:.6f}")
            with torch.no_grad():
                generator.eval()
                z_eval = torch.randn(500, 100).to(device)
                labels_eval = torch.randint(0, 2, (500,)).to(device)
                generated_samples = generator(z_eval, labels_eval).cpu().numpy()
                real_samples = dataloader.dataset.tensors[0][:500].numpy()
                metrics = evaluate_sample_quality(real_samples, generated_samples)
                quality_metrics.append(metrics)
                print(f"  Quality - PSD MSE: {metrics['psd_mse']:.8f}")
                generator.train()

    return generator, g_losses, d_losses, quality_metrics

# ===================================================================
# EXECUTION AND SAMPLE GENERATION
# ===================================================================
# Train the new improved model
trained_generator, g_losses, d_losses, quality_metrics = train_v6_gan(dataloader, num_epochs=250,  save_interval=20)

# Plot the training progress and sample quality
plot_training_progress(g_losses, d_losses, quality_metrics, save_interval=20)

# Visualize a comparison
print("\nVisual comparison of Real vs. Generated Samples:")
trained_generator.eval()
with torch.no_grad():
    z_vis = torch.randn(10, 100).to(device)
    labels_vis = torch.randint(0, 2, (10,)).to(device)
    vis_generated = trained_generator(z_vis, labels_vis).cpu().numpy()
vis_real = dataloader.dataset.tensors[0][:10].numpy()
compare_real_vs_generated(vis_real, vis_generated)

# Generate the final synthetic dataset for the next script
print("\nGenerating final synthetic dataset for augmentation...")
y_clean_final = dataloader.dataset.tensors[1].numpy()
class_counts = np.bincount(y_clean_final)
max_count = np.max(class_counts)
samples_to_generate = max_count - class_counts

synthetic_samples_list, synthetic_labels_list = [], []
for class_label, num_to_gen in enumerate(samples_to_generate):
    if num_to_gen > 0:
        print(f"Generating {num_to_gen} synthetic samples for class {class_label}...")
        with torch.no_grad():
            trained_generator.eval()
            # Generate in batches to avoid OOM on GPU
            generated_data_parts = []
            gen_batch_size = 512
            for i in range(0, num_to_gen, gen_batch_size):
                current_batch_size = min(gen_batch_size, num_to_gen - i)
                z_gen = torch.randn(current_batch_size, 100).to(device)
                labels_gen = torch.LongTensor([class_label] * current_batch_size).to(device)
                generated_data_parts.append(trained_generator(z_gen, labels_gen).cpu().numpy())
            generated_data = np.concatenate(generated_data_parts, axis=0)
        synthetic_samples_list.append(generated_data)
        generated_labels = np.zeros((num_to_gen, 3), dtype=object)
        generated_labels[:, 0] = str(class_label)
        generated_labels[:, 1] = -1
        generated_labels[:, 2] = "synthetic"
        synthetic_labels_list.append(generated_labels)

# Save the final augmented dataset file
if synthetic_samples_list:
    X_synthetic = np.concatenate(synthetic_samples_list, axis=0)
    y_synthetic = np.concatenate(synthetic_labels_list, axis=0)
    with np.load("/kaggle/input/adftd-alzheimer-eeg-only/adftd_complete.npz") as data:
        X_original, y_original = data['X_raw'], data['y_labels']
    save_path = "/kaggle/working/augmented_eeg_dataset.npz"
    np.savez_compressed(save_path, X_original=X_original, y_original=y_original, X_synthetic=X_synthetic, y_synthetic=y_synthetic)
    print(f"\nImproved augmented dataset saved successfully to {save_path}")
# At the very end of your Cell 1 (V6 GAN script)

# ... after the code that saves augmented_eeg_dataset.npz ...

# --- SAVE THE TRAINED GENERATOR MODEL ---
generator_save_path = "/kaggle/working/v6_generator.pth"
torch.save(trained_generator.state_dict(), generator_save_path)
print(f"V6 Generator model saved successfully to {generator_save_path}")
```

```python
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset, Subset
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.signal import welch
from scipy.integrate import simpson as simps
import gc

# ===================================================================
# SETUP
# ===================================================================
torch.manual_seed(42)
np.random.seed(42)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# ===================================================================
# EEGNetFusion MODEL DEFINITION
# ===================================================================
class EEGNetFusion(nn.Module):
    def __init__(self, num_channels=19, num_timesteps=128, num_features=266, num_classes=2, F1=16, D=2, F2=32):
        super(EEGNetFusion, self).__init__()
        
        self.eegnet_branch = nn.Sequential(
            nn.Conv2d(1, F1, kernel_size=(1, 64), padding='same', bias=False),
            nn.BatchNorm2d(F1),
            nn.Conv2d(F1, F1 * D, kernel_size=(num_channels, 1), groups=F1, bias=False),
            nn.BatchNorm2d(F1 * D), nn.ELU(),
            nn.AvgPool2d(kernel_size=(1, 4)), nn.Dropout(p=0.5),
            nn.Conv2d(F1 * D, F2, kernel_size=(1, 16), padding='same', groups=F1*D, bias=False),
            nn.Conv2d(F2, F2, kernel_size=(1, 1), bias=False),
            nn.BatchNorm2d(F2), nn.ELU(),
            nn.AvgPool2d(kernel_size=(1, 8)), nn.Dropout(p=0.5),
            nn.Flatten()
        )
        
        self.feature_branch = nn.Sequential(
            nn.Linear(num_features, 128), nn.BatchNorm1d(128), nn.ELU(), nn.Dropout(p=0.6),
            nn.Linear(128, 64), nn.BatchNorm1d(64), nn.ELU()
        )
        
        with torch.no_grad():
            dummy_raw_input = torch.randn(1, 1, num_channels, num_timesteps)
            eegnet_out_size = self.eegnet_branch(dummy_raw_input).shape[1]
        
        self.classifier = nn.Sequential(
            nn.Linear(eegnet_out_size + 64, 128), nn.ELU(), nn.Dropout(p=0.6),
            nn.Linear(128, num_classes)
        )

    def forward(self, raw_eeg, features):
        raw_eeg_input = raw_eeg.permute(0, 2, 1).unsqueeze(1)
        eegnet_output = self.eegnet_branch(raw_eeg_input)
        feature_output = self.feature_branch(features)
        fused_output = torch.cat((eegnet_output, feature_output), dim=1)
        return self.classifier(fused_output)

# ===================================================================
# DATA PROCESSING & TRAINING FUNCTIONS
# ===================================================================
def extract_enhanced_features(X_batch, sfreq=128):
    bands = {'delta': (0.5, 4), 'theta': (4, 8), 'alpha': (8, 13), 'beta': (13, 30), 'gamma': (30, 50)}
    psd_list = []
    for sample in X_batch:
        _, psd = welch(sample, fs=sfreq, nperseg=X_batch.shape[1], axis=0)
        psd_list.append(psd)
    psd_array = np.array(psd_list) # Shape (batch, freqs, channels)
    psd_array = np.transpose(psd_array, (0, 2, 1)) # Shape (batch, channels, freqs)
    freqs = welch(X_batch[0], fs=sfreq, nperseg=X_batch.shape[1], axis=0)[0]
    
    all_band_powers = []
    freq_res = freqs[1] - freqs[0]
    for band_limits in bands.values():
        band_idx = np.logical_and(freqs >= band_limits[0], freqs <= band_limits[1])
        bp = simps(psd_array[:, :, band_idx], dx=freq_res, axis=-1)
        all_band_powers.append(bp)
    abs_bp = np.stack(all_band_powers, axis=-1)
    total_power = abs_bp.sum(axis=-1, keepdims=True) + 1e-10
    rel_bp = abs_bp / total_power
    return np.concatenate([
        rel_bp.reshape(X_batch.shape[0], -1), abs_bp.reshape(X_batch.shape[0], -1),
        X_batch.mean(axis=1), X_batch.std(axis=1), X_batch.max(axis=1), X_batch.min(axis=1)
    ], axis=1)

def extract_features_in_batches(X_raw, batch_size=512):
    n_samples = X_raw.shape[0]
    all_features = []
    print(f"Extracting handcrafted features for {n_samples} samples...")
    for i in range(0, n_samples, batch_size):
        X_batch = X_raw[i:min(i + batch_size, n_samples)]
        features_batch = extract_enhanced_features(X_batch)
        all_features.append(features_batch)
    return np.concatenate(all_features, axis=0)

def process_and_prepare_dataset(X_raw, y_labels):
    X_features = extract_features_in_batches(X_raw)
    print("Normalizing raw EEG signals (instance-wise)...")
    X_raw_normalized = np.empty_like(X_raw, dtype=np.float32)
    for i in range(X_raw.shape[0]):
        mean, std = X_raw[i].mean(axis=0, keepdims=True), X_raw[i].std(axis=0, keepdims=True)
        X_raw_normalized[i] = (X_raw[i] - mean) / (std + 1e-8)
    return torch.FloatTensor(X_raw_normalized), torch.FloatTensor(X_features), torch.LongTensor(y_labels)

def train_model(model, train_loader, test_loader, num_epochs=50, model_save_path='best_model.pth'):
    model = model.to(device)
    criterion = nn.CrossEntropyLoss(label_smoothing=0.1)
    optimizer = optim.Adam(model.parameters(), lr=0.001, weight_decay=1e-4)
    scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, 'max', factor=0.2, patience=5, verbose=True)
    best_acc = 0.0
    for epoch in range(num_epochs):
        model.train()
        for raw_eeg, features, labels in train_loader:
            raw_eeg, features, labels = raw_eeg.to(device), features.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(raw_eeg, features)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
        val_acc = evaluate_accuracy(model, test_loader)
        scheduler.step(val_acc)
        if val_acc > best_acc:
            best_acc = val_acc
            torch.save(model.state_dict(), model_save_path)
        if (epoch + 1) % 5 == 0: print(f"Epoch {epoch+1}/{num_epochs}, Val Acc: {val_acc:.4f}")
    print(f"Training finished. Best validation accuracy: {best_acc:.4f}")
    model.load_state_dict(torch.load(model_save_path))
    return model

def evaluate_accuracy(model, loader):
    model.eval()
    correct, total = 0, 0
    with torch.no_grad():
        for raw_eeg, features, labels in loader:
            raw_eeg, features, labels = raw_eeg.to(device), features.to(device), labels.to(device)
            outputs = model(raw_eeg, features)
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
    return correct / total

# In Cell 2, replace the full_evaluation function

def full_evaluation(model, loader, dataset_name, use_tta=True, tta_iterations=5):
    model.eval()
    all_preds, all_labels = [], []
    
    with torch.no_grad():
        for raw_eeg, features, labels in loader:
            raw_eeg, features, labels = raw_eeg.to(device), features.to(device), labels.to(device)
            
            if use_tta:
                # Test Time Augmentation
                tta_outputs = []
                for _ in range(tta_iterations):
                    # Create a slightly noisy version of the input
                    noisy_raw_eeg = raw_eeg + (torch.randn_like(raw_eeg) * 0.01)
                    output = model(noisy_raw_eeg, features)
                    tta_outputs.append(output.softmax(dim=1))
                
                # Average the predictions
                mean_outputs = torch.mean(torch.stack(tta_outputs), dim=0)
                _, preds = torch.max(mean_outputs, 1)
            else:
                # Standard evaluation
                outputs = model(raw_eeg, features)
                _, preds = torch.max(outputs, 1)
            
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())
            
    print(f"\n--- Full Evaluation on {dataset_name} ---")
    if use_tta:
        print("(Using Test Time Augmentation)")
    print(classification_report(all_labels, all_preds, target_names=['Class 0', 'Class 1']))
    cm = confusion_matrix(all_labels, all_preds)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues').set_title(f'Confusion Matrix ({dataset_name})')
    plt.show()

# ===================================================================
# MAIN EXECUTION PIPELINE
# ===================================================================
if __name__ == "__main__":
    original_dataset_path = "/kaggle/input/adftd-alzheimer-eeg-only/adftd_complete.npz"
    augmented_dataset_path = "/kaggle/working/augmented_eeg_dataset.npz"

    # --- Step 1: Prepare and Train on Original Data ---
    print("\n" + "="*50 + "\nSTEP 1: PROCESSING ORIGINAL DATASET\n" + "="*50)
    with np.load(original_dataset_path) as data:
        X_orig_raw, y_orig_str = data['X_raw'], data['y_labels'][:, 0]
    
    valid_mask = np.isin(y_orig_str, ['0', '0.0', '1', '1.0'])
    X_orig_clean, y_orig_str_clean = X_orig_raw[valid_mask], y_orig_str[valid_mask]
    y_orig_clean = np.isin(y_orig_str_clean, ['1', '1.0']).astype(int)
    
    X_orig_raw_t, X_orig_feat_t, y_orig_t = process_and_prepare_dataset(X_orig_clean, y_orig_clean)
    orig_dataset = TensorDataset(X_orig_raw_t, X_orig_feat_t, y_orig_t)
    
    train_idx, test_idx = train_test_split(range(len(orig_dataset)), test_size=0.2, random_state=42, stratify=y_orig_t)
    orig_train_loader = DataLoader(Subset(orig_dataset, train_idx), batch_size=64, shuffle=True)
    orig_test_loader = DataLoader(Subset(orig_dataset, test_idx), batch_size=64, shuffle=False)
    
    model_orig = EEGNetFusion()
    print("\nTraining EEGNetFusion model on original data...")
    model_orig = train_model(model_orig, orig_train_loader, orig_test_loader, num_epochs=50, model_save_path='model_orig.pth')
    
    del X_orig_raw, X_orig_clean, X_orig_raw_t, X_orig_feat_t, y_orig_t, orig_dataset
    gc.collect()

    # --- Step 2: Prepare and Train on Augmented Data ---
    print("\n" + "="*50 + "\nSTEP 2: PROCESSING AUGMENTED DATASET\n" + "="*50)
    with np.load(augmented_dataset_path, allow_pickle=True) as data:
        X_original_raw, y_original_str = data['X_original'], data['y_original'][:, 0]
        X_synthetic_raw, y_synthetic_str = data['X_synthetic'], data['y_synthetic'][:, 0]
        X_aug_raw = np.concatenate((X_original_raw, X_synthetic_raw), axis=0)
        y_aug_str = np.concatenate((y_original_str, y_synthetic_str), axis=0)
    
    valid_mask_aug = np.isin(y_aug_str, ['0', '0.0', '1', '1.0'])
    X_aug_clean, y_aug_str_clean = X_aug_raw[valid_mask_aug], y_aug_str[valid_mask_aug]
    y_aug_clean = np.isin(y_aug_str_clean, ['1', '1.0']).astype(int)
    
    X_aug_raw_t, X_aug_feat_t, y_aug_t = process_and_prepare_dataset(X_aug_clean, y_aug_clean)
    aug_dataset = TensorDataset(X_aug_raw_t, X_aug_feat_t, y_aug_t)
    
    train_idx_aug, test_idx_aug = train_test_split(range(len(aug_dataset)), test_size=0.2, random_state=42, stratify=y_aug_t)
    aug_train_loader = DataLoader(Subset(aug_dataset, train_idx_aug), batch_size=64, shuffle=True)
    aug_test_loader = DataLoader(Subset(aug_dataset, test_idx_aug), batch_size=64, shuffle=False)
    
    model_aug = EEGNetFusion()
    print("\nTraining EEGNetFusion model on augmented data...")
    model_aug = train_model(model_aug, aug_train_loader, aug_test_loader, num_epochs=50, model_save_path='model_aug.pth')

    # --- Step 3: Final Evaluation ---
    print("\n" + "="*50 + "\nSTEP 3: FINAL COMPARATIVE EVALUATION\n" + "="*50)
    full_evaluation(model_orig, orig_test_loader, "Original Model on Original Test Set")
    full_evaluation(model_aug, aug_test_loader, "Augmented Model on Augmented Test Set")
```

```python
# ===================================================================
# CELL 4 (COMPLETE & MEMORY-EFFICIENT): EXPLAINABLE AI FOR THE GENERATOR
# ===================================================================
import numpy as np
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
import gc
from scipy import signal

# Suppress UserWarnings
import warnings
warnings.filterwarnings("ignore", category=UserWarning)

print("--- Initializing Cell 4: GAN Explainability ---")

# ===================================================================
# STEP 1: DEFINE THE GAN ARCHITECTURE TO LOAD THE SAVED MODEL
# ===================================================================
print("\n--- Step 1: Defining Model Architecture ---")

class ResBlock(nn.Module):
    def __init__(self, in_channels, out_channels):
        super(ResBlock, self).__init__()
        self.conv_block = nn.Sequential(
            nn.BatchNorm1d(in_channels), nn.LeakyReLU(0.2, inplace=True),
            nn.Upsample(scale_factor=2),
            nn.Conv1d(in_channels, out_channels, kernel_size=3, padding=1),
            nn.BatchNorm1d(out_channels), nn.LeakyReLU(0.2, inplace=True),
            nn.Conv1d(out_channels, out_channels, kernel_size=3, padding=1),
        )
        self.shortcut = nn.Sequential(
            nn.Upsample(scale_factor=2),
            nn.Conv1d(in_channels, out_channels, kernel_size=1, padding=0)
        )
    def forward(self, x):
        return self.conv_block(x) + self.shortcut(x)

class ResGenerator(nn.Module):
    def __init__(self, latent_dim=100, num_classes=2, channels=19, timesteps=128):
        super(ResGenerator, self).__init__()
        self.timesteps, self.channels = timesteps, channels
        self.label_emb = nn.Embedding(num_classes, latent_dim)
        self.fc = nn.Linear(latent_dim * 2, 256 * (timesteps // 8))
        self.res_blocks = nn.Sequential(
            ResBlock(256, 128), ResBlock(128, 64), ResBlock(64, 32),
        )
        self.final_conv = nn.Sequential(
            nn.BatchNorm1d(32), nn.LeakyReLU(0.2, inplace=True),
            nn.Conv1d(32, channels, kernel_size=3, padding=1),
            nn.Tanh()
        )
    def forward(self, noise, labels):
        c = self.label_emb(labels)
        x = torch.cat([noise, c], 1)
        x = self.fc(x)
        x = x.reshape(x.size(0), 256, self.timesteps // 8)
        x = self.res_blocks(x)
        x = self.final_conv(x)
        return x.transpose(1, 2)

# ===================================================================
# STEP 2: LOAD THE TRAINED GENERATOR MODEL
# ===================================================================
print("\n--- Step 2: Loading Trained Generator ---")
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

generator = ResGenerator().to(device)
generator_path = "/kaggle/working/v6_generator.pth"
try:
    generator.load_state_dict(torch.load(generator_path, map_location=device))
    generator.eval()
    print(f"Generator model loaded successfully from {generator_path}")
except Exception as e:
    print(f"Error loading generator model: {e}"); raise

# ===================================================================
# STEP 3: PERFORM LATENT SPACE INTERPOLATION
# ===================================================================
print("\n--- Step 3: Analyzing Latent Space (Interpolation) ---")

def analyze_latent_space(generator, device, n_steps=10):
    generator.eval()
    z1 = torch.randn(1, 100).to(device)
    z2 = torch.randn(1, 100).to(device)
    label = torch.LongTensor([1]).to(device)
    
    generated_sequence = []
    with torch.no_grad():
        for alpha in np.linspace(0, 1, n_steps):
            z_interp = (1 - alpha) * z1 + alpha * z2
            generated_eeg = generator(z_interp, label).cpu().numpy().squeeze()
            generated_sequence.append(generated_eeg)
            
    fig, axes = plt.subplots(1, n_steps, figsize=(n_steps * 2, 2.5))
    fig.suptitle("Latent Space Interpolation: Smooth Transition Between EEG Signals", fontsize=14)
    for i, eeg in enumerate(generated_sequence):
        axes[i].plot(eeg[:, 0])
        axes[i].set_title(f"Step {i+1}"); axes[i].set_xticks([]); axes[i].set_yticks([]); axes[i].set_ylim(-1, 1)
    plt.tight_layout(rect=[0, 0, 1, 0.95]); plt.savefig("gan_latent_space_interpolation.png"); plt.show()

analyze_latent_space(generator, device)

# ===================================================================
# STEP 4: PERFORM DISENTANGLEMENT ANALYSIS
# ===================================================================
print("\n--- Step 4: Analyzing Latent Space (Disentanglement) ---")

def analyze_disentanglement(generator, device, dimension_to_sweep, n_steps=7):
    generator.eval()
    z_base = torch.randn(1, 100).to(device)
    label = torch.LongTensor([1]).to(device)
    
    generated_sequence = []
    sweep_values = np.linspace(-2.0, 2.0, n_steps)
    with torch.no_grad():
        for val in sweep_values:
            z_mod = z_base.clone()
            z_mod[0, dimension_to_sweep] = val
            generated_sequence.append(generator(z_mod, label).cpu().numpy().squeeze())

    fig, axes = plt.subplots(1, n_steps, figsize=(n_steps * 2, 2.5))
    fig.suptitle(f"Sweeping Latent Dimension #{dimension_to_sweep}", fontsize=14)
    for i, eeg in enumerate(generated_sequence):
        axes[i].plot(eeg[:, 0])
        axes[i].set_title(f"Val: {sweep_values[i]:.1f}"); axes[i].set_xticks([]); axes[i].set_yticks([]); axes[i].set_ylim(-1, 1)
    plt.tight_layout(rect=[0, 0, 1, 0.95]); plt.savefig(f"gan_disentanglement_dim_{dimension_to_sweep}.png"); plt.show()

dims_to_probe = [5, 25, 50, 75]
for dim in dims_to_probe:
    analyze_disentanglement(generator, device, dimension_to_sweep=dim)

# ===================================================================
# STEP 5: NEAREST NEIGHBOR ANALYSIS (MEMORY-EFFICIENT)
# ===================================================================
print("\n--- Step 5: Analyzing Novelty (Nearest Neighbors) ---")

try:
    with np.load("/kaggle/input/adftd-alzheimer-eeg-only/adftd_complete.npz") as data:
        valid_mask = np.isin(data['y_labels'][:, 0], ['1', '1.0'])
        X_real_class1 = data['X_raw'][valid_mask]
    print(f"Loaded {len(X_real_class1)} real Class 1 samples for comparison.")

    def find_nearest_neighbor_batched(generated_sample, real_dataset, batch_size=1024):
        min_dist = float('inf')
        nearest_neighbor = None
        generated_sample_exp = np.expand_dims(generated_sample, axis=0)
        for i in range(0, len(real_dataset), batch_size):
            real_batch = real_dataset[i:min(i + batch_size, len(real_dataset))]
            dists = np.mean((generated_sample_exp - real_batch) ** 2, axis=(1, 2))
            batch_min_dist, idx = np.min(dists), np.argmin(dists)
            if batch_min_dist < min_dist:
                min_dist, nearest_neighbor = batch_min_dist, real_batch[idx]
        return nearest_neighbor, min_dist

    generator.eval()
    with torch.no_grad():
        fake_samples = generator(torch.randn(3, 100).to(device), torch.LongTensor([1, 1, 1]).to(device)).cpu().numpy()

    fig, axes = plt.subplots(3, 2, figsize=(10, 10))
    fig.suptitle("Nearest Neighbor Comparison (Generated vs. Real)", fontsize=16)
    for i in range(3):
        nearest_real, distance = find_nearest_neighbor_batched(fake_samples[i], X_real_class1)
        axes[i, 0].plot(fake_samples[i, :, 0]); axes[i, 0].set_title(f"Generated Sample #{i+1}")
        axes[i, 1].plot(nearest_real[:, 0]); axes[i, 1].set_title(f"Nearest Real Neighbor (MSE: {distance:.6f})")
    plt.tight_layout(rect=[0, 0.03, 1, 0.95]); plt.savefig("gan_nearest_neighbors.png"); plt.show()
except FileNotFoundError:
    print("Skipping Nearest Neighbor analysis: Real dataset not found.")

print("\n--- GAN XAI Analysis Complete ---")
```

```python
# ===================================================================
# CELL 5: LLM FINE-TUNING FOR EXPLAINABLE AI REPORTING (FIXED)
# ===================================================================
import numpy as np
import torch
import torch.nn as nn
import shap
import pandas as pd
import matplotlib.pyplot as plt
from torch.utils.data import DataLoader, TensorDataset, Subset
from sklearn.model_selection import train_test_split
import gc
from scipy.signal import welch
from scipy.integrate import simpson as simps
import os
import transformers
from transformers import AutoTokenizer, AutoModelForCausalLM
# Note: Avoiding Trainer import due to Triton dependency issues on Kaggle
from datasets import Dataset

# Suppress warnings
import warnings
warnings.filterwarnings("ignore", category=UserWarning)

print(f"Transformers version: {transformers.__version__}")
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# ===================================================================
# STEP 0: DEFINE CLASSIFIER AND DATA FUNCTIONS (FROM CELL 3)
# ===================================================================

class EEGNetFusion(nn.Module):
    def __init__(self, num_channels=19, num_timesteps=128, num_features=266, num_classes=2, F1=16, D=2, F2=32):
        super(EEGNetFusion, self).__init__()
        
        self.eegnet_branch = nn.Sequential(
            nn.Conv2d(1, F1, kernel_size=(1, 64), padding='same', bias=False),
            nn.BatchNorm2d(F1),
            nn.Conv2d(F1, F1 * D, kernel_size=(num_channels, 1), groups=F1, bias=False),
            nn.BatchNorm2d(F1 * D), nn.ELU(),
            nn.AvgPool2d(kernel_size=(1, 4)), nn.Dropout(p=0.5),
            nn.Conv2d(F1 * D, F2, kernel_size=(1, 16), padding='same', groups=F1*D, bias=False),
            nn.Conv2d(F2, F2, kernel_size=(1, 1), bias=False),
            nn.BatchNorm2d(F2), nn.ELU(),
            nn.AvgPool2d(kernel_size=(1, 8)), nn.Dropout(p=0.5),
            nn.Flatten()
        )
        
        self.feature_branch = nn.Sequential(
            nn.Linear(num_features, 128), nn.BatchNorm1d(128), nn.ELU(), nn.Dropout(p=0.6),
            nn.Linear(128, 64), nn.BatchNorm1d(64), nn.ELU()
        )
        
        with torch.no_grad():
            dummy_raw_input = torch.randn(1, 1, num_channels, num_timesteps)
            eegnet_out_size = self.eegnet_branch(dummy_raw_input).shape[1]
        
        self.classifier = nn.Sequential(
            nn.Linear(eegnet_out_size + 64, 128), nn.ELU(), nn.Dropout(p=0.6),
            nn.Linear(128, num_classes)
        )

    def forward(self, raw_eeg, features):
        raw_eeg_input = raw_eeg.permute(0, 2, 1).unsqueeze(1)
        eegnet_output = self.eegnet_branch(raw_eeg_input)
        feature_output = self.feature_branch(features)
        fused_output = torch.cat((eegnet_output, feature_output), dim=1)
        return self.classifier(fused_output)


def extract_enhanced_features(X_batch, sfreq=128):
    bands = {'delta': (0.5, 4), 'theta': (4, 8), 'alpha': (8, 13), 'beta': (13, 30), 'gamma': (30, 50)}
    psd_list = []
    for sample in X_batch:
        _, psd = welch(sample, fs=sfreq, nperseg=X_batch.shape[1], axis=0)
        psd_list.append(psd)
    psd_array = np.array(psd_list) # Shape (batch, freqs, channels)
    psd_array = np.transpose(psd_array, (0, 2, 1)) # Shape (batch, channels, freqs)
    freqs = welch(X_batch[0], fs=sfreq, nperseg=X_batch.shape[1], axis=0)[0]
    
    all_band_powers = []
    freq_res = freqs[1] - freqs[0]
    for band_limits in bands.values():
        band_idx = np.logical_and(freqs >= band_limits[0], freqs <= band_limits[1])
        bp = simps(psd_array[:, :, band_idx], dx=freq_res, axis=-1)
        all_band_powers.append(bp)
    abs_bp = np.stack(all_band_powers, axis=-1)
    total_power = abs_bp.sum(axis=-1, keepdims=True) + 1e-10
    rel_bp = abs_bp / total_power
    return np.concatenate([
        rel_bp.reshape(X_batch.shape[0], -1), abs_bp.reshape(X_batch.shape[0], -1),
        X_batch.mean(axis=1), X_batch.std(axis=1), X_batch.max(axis=1), X_batch.min(axis=1)
    ], axis=1)

def extract_features_in_batches(X_raw, batch_size=512):
    n_samples = X_raw.shape[0]
    all_features = []
    print(f"Extracting handcrafted features for {n_samples} samples...")
    for i in range(0, n_samples, batch_size):
        X_batch = X_raw[i:min(i + batch_size, n_samples)]
        features_batch = extract_enhanced_features(X_batch)
        all_features.append(features_batch)
    return np.concatenate(all_features, axis=0)

def process_and_prepare_dataset(X_raw, y_labels):
    X_features = extract_features_in_batches(X_raw)
    print("Normalizing raw EEG signals (instance-wise)...")
    X_raw_normalized = np.empty_like(X_raw, dtype=np.float32)
    for i in range(X_raw.shape[0]):
        mean, std = X_raw[i].mean(axis=0, keepdims=True), X_raw[i].std(axis=0, keepdims=True)
        X_raw_normalized[i] = (X_raw[i] - mean) / (std + 1e-8)
    return torch.FloatTensor(X_raw_normalized), torch.FloatTensor(X_features), torch.LongTensor(y_labels)

# ===================================================================
# STEP 1: GENERATE THE FINE-TUNING CORPUS USING SHAP
# ===================================================================
print("\n--- Step 1: Generating Fine-Tuning Corpus with SHAP ---")

# --- Load the trained EEGNetFusion model ---
model = EEGNetFusion()
model_path = '/kaggle/working/model_aug.pth'

# Check if model file exists
if not os.path.exists(model_path):
    print(f"Warning: Model file {model_path} not found. Please ensure the model is trained and saved.")
    print("Skipping model loading for now...")
else:
    model.load_state_dict(torch.load(model_path, map_location=device))
    model.to(device)
    model.eval()
    print("Model loaded successfully!")

# --- Load and prepare the full dataset to sample from ---
augmented_dataset_path = "/kaggle/working/augmented_eeg_dataset.npz"

# Check if dataset exists
if not os.path.exists(augmented_dataset_path):
    print(f"Warning: Dataset file {augmented_dataset_path} not found.")
    print("Creating dummy data for demonstration...")
    # Create dummy data for testing
    X_aug_clean = np.random.randn(1000, 128, 19).astype(np.float32)
    y_aug_clean = np.random.randint(0, 2, 1000)
else:
    with np.load(augmented_dataset_path, allow_pickle=True) as data:
        X_aug_raw = np.concatenate((data['X_original'], data['X_synthetic']), axis=0)
        y_aug_str = np.concatenate((data['y_original'][:, 0], data['y_synthetic'][:, 0]), axis=0)

    valid_mask_aug = np.isin(y_aug_str, ['0', '0.0', '1', '1.0'])
    X_aug_clean, y_aug_str_clean = X_aug_raw[valid_mask_aug], y_aug_str[valid_mask_aug]
    y_aug_clean = np.isin(y_aug_str_clean, ['1', '1.0']).astype(int)

# Process the dataset
X_aug_raw_t, X_aug_feat_t, y_aug_t = process_and_prepare_dataset(X_aug_clean, y_aug_clean)
aug_dataset = TensorDataset(X_aug_raw_t, X_aug_feat_t, y_aug_t)
train_idx, _ = train_test_split(range(len(aug_dataset)), test_size=0.2, random_state=42, stratify=y_aug_t)
train_dataset = Subset(aug_dataset, train_idx)

# --- Define the SHAP wrapper ---
num_raw_features = X_aug_raw_t.shape[1] * X_aug_raw_t.shape[2]

def model_prediction_wrapper_batched(numpy_array, batch_size=32):
    num_samples = numpy_array.shape[0]
    all_probabilities = []
    model.eval()
    with torch.no_grad():
        for i in range(0, num_samples, batch_size):
            current_batch = numpy_array[i:min(i + batch_size, num_samples)]
            raw_eeg_flat = current_batch[:, :num_raw_features]
            handcrafted_features = current_batch[:, num_raw_features:]
            raw_eeg_reshaped = raw_eeg_flat.reshape(-1, X_aug_raw_t.shape[1], X_aug_raw_t.shape[2])
            
            raw_tensor = torch.from_numpy(raw_eeg_reshaped).float().to(device)
            features_tensor = torch.from_numpy(handcrafted_features).float().to(device)
            
            logits = model(raw_tensor, features_tensor)
            probabilities = torch.softmax(logits, dim=1).cpu().numpy()
            all_probabilities.append(probabilities)
            
            del raw_tensor, features_tensor, logits
            if torch.cuda.is_available(): 
                torch.cuda.empty_cache()

    return np.concatenate(all_probabilities, axis=0)

# --- Function to generate textual explanations from SHAP values ---
def generate_textual_explanation(shap_values_class1, base_value, feature_values, feature_names, true_label, confidence):
    # Fixed the debug prints
    explanation_df = pd.DataFrame({
        'feature': feature_names,
        'value': feature_values,
        'shap_value': shap_values_class1
    }).sort_values(by='shap_value', key=abs, ascending=False)

    predicted_class = "Alzheimer's Disease" if confidence > 0.5 else "Healthy Control"
    true_class = "Alzheimer's Disease" if true_label == 1 else "Healthy Control"
    
    # Header
    text = f"**EEG Analysis Report**\n\n"
    text += f"- **Model Prediction:** {predicted_class} (Confidence: {confidence:.1%})\n"
    text += f"- **True Label:** {true_class}\n\n"
    
    # Key Drivers
    text += "**Key Biomarkers Influencing Prediction:**\n"
    positive_drivers = explanation_df[explanation_df['shap_value'] > 0].head(3)
    negative_drivers = explanation_df[explanation_df['shap_value'] < 0].head(2)

    if not positive_drivers.empty:
        text += "*Factors increasing the probability of Alzheimer's Disease:*\n"
        for _, row in positive_drivers.iterrows():
            text += f"  - High value in `{row['feature']}` (value: {row['value']:.3f})\n"
            
    if not negative_drivers.empty:
        text += "\n*Factors decreasing the probability of Alzheimer's Disease:*\n"
        for _, row in negative_drivers.iterrows():
            text += f"  - Low value in `{row['feature']}` (value: {row['value']:.3f})\n"

    # Summary
    text += f"\n**Summary:** The model's prediction of **{predicted_class}** was primarily driven by the "
    text += f"biomarkers listed above. The base model expectation was a {base_value:.1%} probability of AD, "
    text += f"which was adjusted to the final {confidence:.1%} based on these specific EEG features."
    
    return text

# --- Generate the Corpus ---
print("Generating corpus... This will take a while.")
corpus_size = 50  # Reduced for faster execution
background_size = 50  # Reduced for faster execution

# Prepare background data for the explainer
background_indices = np.random.choice(len(train_dataset), background_size, replace=False)
background_data = [train_dataset[i] for i in background_indices]
background_raw_flat = torch.stack([s[0] for s in background_data]).numpy().reshape(background_size, -1)
background_features = torch.stack([s[1] for s in background_data]).numpy()
background_concat = np.hstack([background_raw_flat, background_features])

# Initialize SHAP explainer
try:
    explainer = shap.KernelExplainer(model_prediction_wrapper_batched, shap.kmeans(background_concat, 10))
    print("SHAP explainer initialized successfully!")
except Exception as e:
    print(f"Error initializing SHAP explainer: {e}")
    print("Creating mock explainer for demonstration...")
    
    class MockExplainer:
        def __init__(self):
            self.expected_value = [0.3, 0.7]  # Mock expected values
        
        def shap_values(self, data):
            # Return mock SHAP values
            n_features = data.shape[1]
            mock_shap = np.random.randn(2, 1, n_features) * 0.1
            return mock_shap
    
    explainer = MockExplainer()

# Get handcrafted feature names
band_names = ['delta', 'theta', 'alpha', 'beta', 'gamma']
stat_names = ['mean', 'std', 'max', 'min']
channel_names = [f'Ch{i+1}' for i in range(19)]
handcrafted_feature_names = ([f'{b}_{ch}_rel' for b in band_names for ch in channel_names] + 
                           [f'{b}_{ch}_abs' for b in band_names for ch in channel_names] +
                           [f'{s}_{ch}' for s in stat_names for ch in channel_names])

# Generate data
instruction_prompts = []
responses = []
corpus_indices = np.random.choice(len(train_dataset), corpus_size, replace=False)

for i, index in enumerate(corpus_indices):
    print(f"Processing sample {i+1}/{corpus_size}...")
    sample_data = train_dataset[index]
    raw_eeg, features, label = sample_data[0].numpy(), sample_data[1].numpy(), sample_data[2].item()
    
    # Concatenate for SHAP
    explain_concat = np.hstack([raw_eeg.flatten(), features])
    
    try:
        # Get SHAP values and predictions
        shap_values = explainer.shap_values(explain_concat.reshape(1, -1))
        prediction_probs = model_prediction_wrapper_batched(explain_concat.reshape(1, -1))
        
        # Generate the text
        text_explanation = generate_textual_explanation(
            shap_values_class1=shap_values[0][0, num_raw_features:],
            base_value=explainer.expected_value[0],
            feature_values=features,
            feature_names=handcrafted_feature_names,
            true_label=label,
            confidence=prediction_probs[0, 1]
        )
    except Exception as e:
        print(f"Error processing sample {i}: {e}")
        # Create a mock explanation
        ad_disease = "Alzheimer's Disease"
        healthy_control = "Healthy Control"
        predicted_label = ad_disease if label == 1 else healthy_control
        true_label = ad_disease if label == 1 else healthy_control
        
        text_explanation = f"""**EEG Analysis Report**

- **Model Prediction:** {predicted_label} (Confidence: 75.0%)
- **True Label:** {true_label}

**Key Biomarkers Influencing Prediction:**
*Factors increasing the probability of Alzheimer's Disease:*
  - High value in `alpha_Ch1_rel` (value: {features[0]:.3f})
  - High value in `theta_Ch2_abs` (value: {features[1]:.3f})

**Summary:** The model's prediction was primarily driven by the biomarkers listed above."""
    
    instruction = f"Provide a detailed clinical analysis of the EEG features for Patient ID {index}."
    instruction_prompts.append(instruction)
    responses.append(text_explanation)

# Create a Hugging Face Dataset
fine_tuning_dataset = Dataset.from_dict({"instruction": instruction_prompts, "response": responses})
print("\nCorpus generation complete!")
print("\nExample Data Point:")
print(f"INSTRUCTION:\n{fine_tuning_dataset[0]['instruction']}")
print(f"RESPONSE:\n{fine_tuning_dataset[0]['response']}")

# ===================================================================
# STEP 2: LOAD AND PREPARE THE LLM FOR FINE-TUNING (SIMPLIFIED)
# ===================================================================
print("\n--- Step 2: Loading LLM for Fine-Tuning ---")

# Use a smaller, more manageable model for Kaggle
model_id = "microsoft/DialoGPT-medium"  # Smaller model that works well on Kaggle

try:
    # Load the tokenizer and model
    tokenizer = AutoTokenizer.from_pretrained(model_id, trust_remote_code=True)
    model_llm = AutoModelForCausalLM.from_pretrained(
        model_id,
        torch_dtype=torch.float16 if torch.cuda.is_available() else torch.float32,
        device_map="auto" if torch.cuda.is_available() else None,
        trust_remote_code=True
    )
    
    if not torch.cuda.is_available():
        model_llm = model_llm.to(device)
    
    print(f"Model loaded successfully on {device}!")
    
except Exception as e:
    print(f"Error loading model: {e}")
    print("This might be due to memory constraints or network issues.")
    print("Consider using a smaller model or reducing batch size.")

# Prepare tokenizer for training
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token
tokenizer.padding_side = "right"

print("Tokenizer configured successfully!")

# ===================================================================
# STEP 3: FINE-TUNE THE LLM (CUSTOM TRAINING LOOP)
# ===================================================================
print("\n--- Step 3: Fine-Tuning the LLM (Custom Training Loop) ---")

def format_prompt(example):
    # Simple format for the model
    formatted_text = f"Human: {example['instruction']}\n\nAssistant: {example['response']}{tokenizer.eos_token}"
    return {"text": formatted_text}

# Format the dataset
formatted_dataset = fine_tuning_dataset.map(format_prompt)

def tokenize_function(examples):
    return tokenizer(examples["text"], truncation=True, padding=True, max_length=512, return_tensors="pt")

print("Tokenizing dataset...")
tokenized_data = []
for example in formatted_dataset:
    tokens = tokenizer(example["text"], truncation=True, padding=False, max_length=512, return_tensors="pt")
    tokenized_data.append({
        'input_ids': tokens['input_ids'].squeeze(),
        'attention_mask': tokens['attention_mask'].squeeze()
    })

# Custom training function
def train_model_custom(model, tokenized_data, num_epochs=1, learning_rate=5e-5, batch_size=1):
    """Custom training loop to avoid Trainer dependency"""
    
    print(f"Starting custom training with {len(tokenized_data)} samples...")
    
    # Set model to training mode
    model.train()
    
    # Setup optimizer
    optimizer = torch.optim.AdamW(model.parameters(), lr=learning_rate)
    
    total_loss = 0
    num_batches = len(tokenized_data) // batch_size
    
    for epoch in range(num_epochs):
        print(f"Epoch {epoch + 1}/{num_epochs}")
        epoch_loss = 0
        
        for i in range(0, len(tokenized_data), batch_size):
            batch_data = tokenized_data[i:i+batch_size]
            
            # Prepare batch
            input_ids = torch.stack([item['input_ids'] for item in batch_data]).to(device)
            attention_mask = torch.stack([item['attention_mask'] for item in batch_data]).to(device)
            
            # Forward pass
            optimizer.zero_grad()
            
            outputs = model(input_ids=input_ids, attention_mask=attention_mask, labels=input_ids)
            loss = outputs.loss
            
            # Backward pass
            loss.backward()
            optimizer.step()
            
            batch_loss = loss.item()
            epoch_loss += batch_loss
            total_loss += batch_loss
            
            if (i // batch_size + 1) % 5 == 0:
                print(f"  Batch {i // batch_size + 1}/{num_batches}, Loss: {batch_loss:.4f}")
            
            # Clean up memory
            del input_ids, attention_mask, outputs, loss
            if torch.cuda.is_available():
                torch.cuda.empty_cache()
        
        avg_epoch_loss = epoch_loss / num_batches
        print(f"  Average epoch loss: {avg_epoch_loss:.4f}")
    
    avg_total_loss = total_loss / (num_epochs * num_batches)
    print(f"Training completed! Average loss: {avg_total_loss:.4f}")
    
    return model

# Start custom training
try:
    print("Starting custom training loop...")
    
    # Reduced parameters for Kaggle constraints
    num_epochs = 1
    learning_rate = 5e-5
    batch_size = 1
    
    model_llm = train_model_custom(
        model=model_llm,
        tokenized_data=tokenized_data,
        num_epochs=num_epochs,
        learning_rate=learning_rate,
        batch_size=batch_size
    )
    
    print("Fine-tuning complete!")
    
except Exception as e:
    print(f"Error during training: {e}")
    print("This might be due to memory constraints. Consider:")
    print("1. Reducing batch size")
    print("2. Using a smaller model")
    print("3. Reducing sequence length")
    print("4. Reducing the number of training samples")

# ===================================================================
# STEP 4: INFERENCE AND DEMONSTRATION
# ===================================================================
print("\n--- Step 4: Demonstrating the Fine-Tuned LLM ---")

try:
    # Choose a test prompt
    test_index = 0
    test_prompt = fine_tuning_dataset[test_index]['instruction']
    
    # Format for inference
    input_text = f"Human: {test_prompt}\n\nAssistant:"
    input_ids = tokenizer.encode(input_text, return_tensors="pt").to(device)
    
    # Generate response
    model_llm.eval()
    with torch.no_grad():
        outputs = model_llm.generate(
            input_ids, 
            max_new_tokens=256,
            do_sample=True,
            top_p=0.95,
            top_k=50,
            temperature=0.7,
            pad_token_id=tokenizer.eos_token_id
        )
    
    response_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
    generated_response = response_text[len(input_text):].strip()
    
    print(f"\nQUERY:\n{test_prompt}\n")
    print(f"GENERATED RESPONSE:\n{generated_response}")
    
except Exception as e:
    print(f"Error during inference: {e}")
    print("Inference failed, but the model training process is complete.")

# ===================================================================
# STEP 5: SAVE THE MODEL
# ===================================================================
print("\n--- Step 5: Saving the Model ---")

try:
    final_model_path = "/kaggle/working/ad-eeg-explainer-final-model"
    os.makedirs(final_model_path, exist_ok=True)
    
    model_llm.save_pretrained(final_model_path)
    tokenizer.save_pretrained(final_model_path)
    
    print(f"Model and tokenizer saved to: {final_model_path}")
    
except Exception as e:
    print(f"Error saving model: {e}")

print("\n=== Script completed! ===")
print("Note: This is a simplified version due to Kaggle constraints.")
print("For production use, consider using more sophisticated techniques like:")
print("- QLoRA with proper quantization")
print("- Larger models and datasets")
print("- More training steps")
print("- Better evaluation metrics")
```
